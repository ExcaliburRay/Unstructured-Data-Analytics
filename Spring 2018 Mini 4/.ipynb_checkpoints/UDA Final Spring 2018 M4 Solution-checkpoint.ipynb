## Get the 20 Newsgroups data 
First, load the data.  Note that this data has been split into two sets: one for training (exploratory data analysis and model selection) and the other one for testing (performance evaluation). The split between the train and test set is based upon a messages posted before and after a specific date. (http://scikit-learn.org/stable/datasets/twenty_newsgroups.html)

```python
from sklearn.datasets import fetch_20newsgroups
import numpy as np

cats = ['comp.sys.ibm.pc.hardware', 'comp.sys.mac.hardware', 'rec.sport.baseball', 'sci.space', 'talk.politics.misc']

train_dataset = fetch_20newsgroups(subset='train', shuffle=True, random_state=95865, remove=('headers', 'footers', 'quotes'), categories = cats)
train_documents = train_dataset.data
train_labels = np.array(train_dataset.target)
num_train = len(train_documents)

test_dataset = fetch_20newsgroups(subset='test', shuffle=True, random_state=95865, remove=('headers', 'footers', 'quotes'), categories = cats)
test_documents = test_dataset.data
test_labels = np.array(test_dataset.target)
num_test = len(test_documents)

all_documents = train_documents + test_documents
all_labels = np.array(list(train_labels) + list(test_labels))

label_names = list(train_dataset.target_names)
short_label_names = ['PC', 'Mac', 'Baseball', 'Space', 'Politics']
```

## Approach 1: Baseline (Word Frequencies) [10 points]

In the first approach, you will construct a matrix of word frequencies and kNN for prediction.  

### Construct the document matrix [5 points]

   - Use the `CountVectorizer` from scikit-learn.
   - Set `min_df=2`; this drops words that don't occur in at least 2 documents.
   - Set `stop_words="english"` and `max_df=0.95` to filter out stop-words.
   - Set `max_features=1000` to only consider the 1,000 most frequent terms in the corpus.

```python
from sklearn.feature_extraction.text import CountVectorizer

no_features = 1000
```

```python
tf_vectorizer = CountVectorizer(max_df=0.95, min_df=2, max_features=no_features, stop_words='english')
tf = tf_vectorizer.fit_transform(train_documents)
tf_feature_names = tf_vectorizer.get_feature_names()
```

```python
analyze = tf_vectorizer.build_analyzer()
train_documents = [analyze(doc) for doc in train_documents]
```

### Model Selection to find the optimal number of neighbors (k) [3 points]
* Train a kNN model to predict the newsgroup category.  
* Use the documents' word frequencies as the features in this model (the x's).  
* Use 5 fold Cross Validation to find the number of neighbors which gives the highest accuracy in predicting the newsgroup.
* Select between k = 1, 2, 5, 10, 25, 100 or 1000


```python
from sklearn.model_selection import KFold
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
import numpy as np
```

```python
num_folds = 5
k_fold = KFold(num_folds, shuffle=True, random_state = 95865)
tf_array = tf.toarray()

# K-NN
param_values = [1, 2, 5, 10, 25, 100, 1000]

cross_val_scores = []

for p in param_values:
    # your code to train and score the training data here
    print("k = " + str(p))
    fold_scores = []
    for k,(train_index,val_index) in enumerate(k_fold.split(tf_array,train_labels)):
        neig = KNeighborsClassifier(n_neighbors = p)
        neig.fit(tf_array[train_index], train_labels[train_index])
        ypred = neig.predict(tf_array[val_index])
        yval = train_labels[val_index]
        
        acc = accuracy_score(yval,ypred)
        
        fold_scores.append(acc)
        
        print("\t[fold {0}] k: {1:.5f}, accuracy: {2:.5f}".format(k, p, acc))

    cross_val_scores.append(np.mean(fold_scores))
    
best_k_index = np.argmax(cross_val_scores)
best_k = param_values[best_k_index]
print("Best K: " + str(best_k) + " with score: " + str(cross_val_scores[best_k_index]))
```

```python
from sklearn.decomposition import LatentDirichletAllocation
topic_models = []
num_topics = [2,3,4,5,6,7,8,9,10]
for k in num_topics:
    print("Applying LDA for k=%d ..." % k )

    model = LatentDirichletAllocation(n_components=k, max_iter=1,
                                      learning_method='online', random_state=95865).fit(all_tf)

    W = model.fit_transform(all_tf)
    H = model.components_
    
    
    
    topic_models.append((k,W,H))
```

### Function to compute topic coherence [10 points]
* We will use coherence to determine the optimal number of topics.  Coherence has similarities to pointwise mutual information (PMI), and research has found that it is a useful metric for selecting the optimal number of topics.  We discussed coherence in class, but have not computed it until now.  


* Coherence is computed as follows:
* Let $D(v)$ be the document frequency of word v (i.e., the number of documents with least one token of type v), and let $D(v, v')$ be the co-document frequency of words v and v' (i.e., the number of documents containing both v and v', not necessarily in sequence):
* Let $V^{(t)} = (v_1^{(t)}, v_2^{(t)}, ... , v_M^{(t)})$ be a list of the M most probable words in topic t.


$$
    \textrm{Coherence for topic t} = \sum_{m=2}^{M}\sum_{l=1}^{m-1}\textrm{score}(v_m^{(t)},v_l^{(t)})\\
    \textrm{score}(v_m^{(t)},v_l^{(t)}) = \frac{D(v_m^{(t)}, v_l^{(t)}) + 1}{D(v_l^{(t)})}\\
    \textrm{Overall coherence of LDA model} = \textrm{Mean coherence of all topics in LDA model}
$$


#### Write a function to compute the overall topic coherence for a LDA model.
* In your function, let M = 10.
* Hint 1: You will need to compute document and co-document frequencies first.
* Hint 2: The inputs to the coherence function should be the document frequencies, the co-document frequencies, and the most probable words for each topic. The output should be the overall coherence of the LDA model.
* Hint 3: It may be helpful to look at the notes from the PMI recitation, which can be found here:  https://gist.github.com/emaadmanzoor/1d06e0751a3f7d39bc6814941b37531d

```python
%%time
from collections import Counter
from itertools import combinations
import re

# compute document frequencies
cx = Counter()
cxy = Counter()
all_words = set(all_tf_feature_names)
for i, text in enumerate(all_documents):
    words = set(text).intersection(all_words)
    for x in words:
        cx[x] += 1
    for x, y in map(sorted, combinations(words, 2)):
        cxy[(x, y)] += 1
```

```python
def coherence(term_rankings, cx, cxy):
    overall_coherence = 0.0
    for topic_index in range(len(term_rankings)):
        # check each pair of terms
        pair_scores = []
        for pair in combinations( term_rankings[topic_index], 2 ):
            pair_scores.append(float(cxy[pair[0], pair[1]] + 1)/cx[pair[0]])
        # get the mean for all pairs in this topic
        topic_score = sum(pair_scores) / len(pair_scores)
        overall_coherence += topic_score
    # get the mean score across all topics
    return overall_coherence / len(term_rankings)
```

### Compute and plot the topic coherence for each k [10 points]

```python
import numpy as np
def get_descriptor(all_terms, H, topic_index, top):
    # reverse sort the values to sort the indices
    top_indices = np.argsort( H[topic_index,:] )[::-1]
    # now get the terms corresponding to the top-ranked indices
    top_terms = []
    for term_index in top_indices[0:top]:
        top_terms.append( all_terms[term_index] )
    return top_terms
```

```python
from itertools import combinations
k_values = []
coherences = []
for (k,W,H) in topic_models:
    # Get all of the topic descriptors - the term_rankings, based on top 10 terms
    term_rankings = []
    for topic_index in range(k):
        term_rankings.append( get_descriptor(all_tf_feature_names, H, topic_index, 10))
    # Now calculate the coherence based on our Word2vec model
    k_values.append( k )
    coherences.append(coherence(term_rankings, cx, cxy))
    print("K=%02d: Coherence=%.4f" % (k, coherences[-1]))
```

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt

k_values = num_topics
fig = plt.figure(figsize=(13,7))
ax = plt.plot( k_values, coherences )
plt.xticks(k_values)
plt.xlabel("Number of Topics")
plt.ylabel("Mean Coherence")
plt.scatter( k_values, coherences, s=120)
ymax = max(coherences)
xpos = coherences.index(ymax)
best_k = k_values[xpos]
plt.annotate( "k=%d" % best_k, xy=(best_k, ymax), xytext=(best_k, ymax), textcoords="offset points", fontsize=16)
plt.grid()
plt.show()
```

### Show the Confusion Matrices for Each Approach [5 points]

* Please write code to display the confusion matrices for each of the 3 approaches.
* Please normalize these matrices so that they show the percent of articles from each true newsgroup that were classified as each of the 5 categories.  

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')  # prettier plots

from sklearn.metrics import confusion_matrix
import itertools

def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
from sklearn.metrics import confusion_matrix

conf_matrix = confusion_matrix(test_labels, base_ypred)
plot_confusion_matrix(conf_matrix, classes=short_label_names, title='Baseline Confusion matrix', normalize=True)
```

```python
topic_conf_matrix = confusion_matrix(test_labels, topic_ypred)
plot_confusion_matrix(topic_conf_matrix, classes=short_label_names, title='Topic Model Confusion matrix', normalize=True)
```

```python
tsne_conf_matrix = confusion_matrix(test_labels, tsne_ypred)
plot_confusion_matrix(tsne_conf_matrix, classes=short_label_names, normalize = True, title='t-SNE Confusion matrix')
```

#### 1. What insights about the benefits of the various approaches can you gain from the confusion matrices that were not obvious by just looking at the accuracy metric?

Looking at the confusion matrices provides insight into the prediction errors.  For example, we see that after dimensionality reduction is applied, the models have the most difficulty distinguishing between the PC and Mac newsgroups.  The confusion matrices also show how each model performs for each class, while the accuracy metric just shows the average performance across classes.

#### 2. What is the easiest newsgroup to predict?

In a confusion matrix, the diagonal elements give the class accuracies.  Since the accuracy is highest for the baseball group in all of the models, the baseball newsgroup is the easiest to predict. 

#### 3. Why did we perform dimensionality reduction over the entire corpus?  Why doesn't this violate principles of cross validation?

We need for the training and test data to both have the same representation, so dimensionality reduction must be applied to both sets. 

The procedure that we followed in this exam is a simplification that *does* violate principles of cross validation, however.  When doing dimensionality reduction as in approaches 2 and 3, the dimensionality should *not* look at the test data. Dimensionality reduction methods typically depend on all the inputs and so by using test data when doing the dimensionality reduction, you are contaminating your training procedure with test information.

In practice, you would want to perform dimensionality reduction on your training set only.  The question then becomes how do you project down the test data. Usually this is done in 1 of 2 ways: either the dimensionality reduction method gives you a way to do the projection (for example, PCA gives you the transformation matrix for doing the projection), or if the dimensionality reduction method you use does *not* have a way to project new data points down to the lower-dimensional space, then you need to use an "out-of-sample extension" method. There are a number of ways to do this but at a high-level it involves learning a multi-output regression function that approximates the projection.  t-SNE is a dimensionality reduction method that requires an out-of-sample extension method.
