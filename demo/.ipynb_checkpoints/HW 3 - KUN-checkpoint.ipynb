````python
import glob

ham = glob.glob('./email-date/ham/*.txt')
spam = glob.glob('./email-date/spam/*.txt')

print("The number of ham mail is {}".format(len(ham)))
print("The number of spam mail is {}".format(len(spam)))
````

### Construct the documents [4 points]
 
   - Provided below is a function that returns a document present in a file given a fileName.
   - The function performs some preprocessing to (1) remove punctuation, (2),(3) remove whitespace and (4) lowercase all words.
   - Use this function to construct a list of documents.
   - Also construct a list of document labels containing `1` for spam and `0` for ham.

````python
import re
import string
import codecs

def makeWordList(path):
    
    with codecs.open(path, "r", encoding='utf-8', errors='ignore') as f:
        corpus_text = f.read()

    for c in string.punctuation:
        corpus_text = corpus_text.replace(c, "")  # -- (1)
    
    text = re.sub(r'\S*\d\S*','',corpus_text) # -- (2)
    text = re.sub(r'[^\w\s]','',text)         # -- (3)
    
    text = text.lower().split()           # -- (4)         
    
    li = []
    for token in text:
        li.append(token)

    return " ".join(li)
````

````python
import numpy as np

document_content = []
document_labels = []

for ham_email in ham:
    document_content.append(makeWordList(ham_email))
    document_labels.append(0)

for spam_email in spam:
    document_content.append(makeWordList(spam_email))
    document_labels.append(1)
    
document_labels = np.array(document_labels)
````

### Construct the document matrix `X` as a matrix of word frequencies [5 points]

   - Use the `CountVectorizer` from scikit-learn.
   - Set `min_df=50`; this drops words that don't occur in at least 50 documents.
   - Set `stop_words="english"` and `max_df=0.8` to filter out stop-words.
   - Print the size of the vocabulary (number of unique words)

````python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(max_df=0.8, min_df=50, stop_words='english')
X = vectorizer.fit_transform(document_content)

size_of_vocabulary = X.shape[1]
print(f"The size of vocabulary is {size_of_vocabulary}")
````

### k-NN, SVM, random forest: Hyperparameter selection [15 points]

Now that you have your documents and labels as training data, you can perform 5-fold cross-validation to select the hyperparameters for different learning algorithms.

The hyperparameter with the best performance averaged across 5 folds is chosen. Use the **weighted F1-score** as the evaluation metric.

   - k-NN: Select `k` from a range of values of your choice.
   - SVM: (SVC) Select `C` from a range of your choice, use any kernel that performs well.
   - Random forest: Select `n_estimators` **and** `max_depth` from a grid of your choice.

Store each chosen hyperparameter as `best_k`, `best_C`, `best_n_estimators` **and** `best_max_depth` respectively.

Provided is some seed code for cross-validation that you may modify and reuse. Do not use the cross-validations score or grid-search functions from scikit-learn (you may use `KFold`).

#### import common libary ####

````python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier

from sklearn.model_selection import KFold
from sklearn.metrics import f1_score
````

#### Calculate best_k ####

````python
best_k = None
arg_max = None
max_cross_val_score = -np.inf

indices = range(size_of_vocabulary)
num_folds = 5
k_fold = KFold(n_splits=num_folds, shuffle=True)

for k in range(1, 50):
    fold_scores = []
    
    for train_indices, val_indices in k_fold.split(indices):
        classifier = KNeighborsClassifier(n_neighbors=k)
        classifier.fit(X[train_indices], document_labels[train_indices])
        predicted_val_labels = classifier.predict(X[val_indices])
        
        fold_score = f1_score(document_labels[val_indices],predicted_val_labels, average='weighted')
        fold_scores.append(fold_score)
        
    cross_val_score = np.mean(fold_scores)
    print('k:', k, 'cross validation score:', cross_val_score)
    
    if cross_val_score > max_cross_val_score:
        max_cross_val_score = cross_val_score
        arg_max = k
            
best_k = arg_max
            
print("Best k is: ", best_k)
````

#### Calculate best_C ####

````python
best_C = None
arg_max = None
max_cross_val_score = -np.inf

indices = range(size_of_vocabulary)
num_folds = 5
k_fold = KFold(n_splits=num_folds, shuffle=True, random_state=0)

for C in np.logspace(-4, 2, 3):
    fold_scores = []
    
    for train_indices, val_indices in k_fold.split(indices):
        classifier = SVC(C=C)
        classifier.fit(X[train_indices], document_labels[train_indices])
        predicted_val_labels = classifier.predict(X[val_indices])
        
        fold_score = f1_score(document_labels[val_indices],predicted_val_labels, average='weighted')
        fold_scores.append(fold_score)

    cross_val_score = np.mean(fold_scores)
    print('C:', C, 'fold scores:', cross_val_score)
    if cross_val_score > max_cross_val_score:
        max_cross_val_score = cross_val_score
        arg_max = C
            
best_C = arg_max
print("Best C is: ", best_C)
````

#### Calculate best_n_estimators and best_max_depth ####

````python
best_n_estimators = None
best_max_depth = None
estimators_max = None
depth_max = None
max_cross_val_score = -np.inf

indices = range(size_of_vocabulary)
num_folds = 5
k_fold = KFold(n_splits=num_folds, shuffle=True, random_state=0)

for estimators in [80, 90, 100, 110, 120, 130]:
    for depth in [60, 70, 80, 90, 100]:
        fold_scores = []
        
        for train_indices, val_indices in k_fold.split(indices):
            classifier = RandomForestClassifier(n_estimators = estimators, max_depth = depth, random_state = 0)
            classifier.fit(X[train_indices], document_labels[train_indices])
            predicted_val_labels = classifier.predict(X[val_indices])

            fold_score = f1_score(document_labels[val_indices],predicted_val_labels, average='weighted')
            fold_scores.append(fold_score)
            
        cross_val_score = np.mean(fold_scores)
        print('estimators:', estimators, 'depth:', depth, 'fold scores:', cross_val_score)
        if cross_val_score > max_cross_val_score:
            max_cross_val_score = cross_val_score
            estimators_max = estimators
            depth_max = depth
            
best_n_estimator = estimators_max
best_max_depth = depth_max
print("Best n estimators is: ", best_n_estimator, "Best max depth is: ", best_max_depth)
````

````python
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve
from sklearn.metrics import auc

# load testing data
test = glob.glob('./email-date/testing/*.txt')
test_document_content = []
test_document_labels = []

# add label and processing the data
for test_email in test:
    test_document_content.append(makeWordList(test_email))
    if test_email.split(".")[-2] == "ham":
        label = 0
    else:
        label = 1
    test_document_labels.append(label)

test_document_labels = np.array(test_document_labels)
test_X = vectorizer.transform(test_document_content)

# best knn
best_knn_classifier = KNeighborsClassifier(n_neighbors=best_k)
best_knn_classifier.fit(X, document_labels)
predicted_knn_test_labels = best_knn_classifier.predict(test_X)
knn_prob = best_knn_classifier.predict_proba(test_X)[:, 1]

# best svm
best_svm_classifier = SVC(C=best_C, probability=True)
best_svm_classifier.fit(X, document_labels)
predicted_svm_test_labels = best_svm_classifier.predict(test_X)
svm_prob = best_svm_classifier.predict_proba(test_X)[:, 1]

# best random forest
best_rf_classifier = RandomForestClassifier(n_estimators = best_n_estimator, max_depth = best_max_depth)
best_rf_classifier.fit(X, document_labels)
predicted_rf_test_labels = best_rf_classifier.predict(test_X)
rf_prob = best_rf_classifier.predict_proba(test_X)[:, 1]

#Computer ROC curve and ROC area
knn_fpr, knn_tpr, threshold = roc_curve(test_document_labels, knn_prob)
knn_roc_auc = auc(knn_fpr, knn_tpr)

svm_fpr, svm_tpr, threshold = roc_curve(test_document_labels, svm_prob)
svm_roc_auc = auc(svm_fpr, svm_tpr)

rf_fpr, rf_tpr, threshold = roc_curve(test_document_labels, rf_prob)
rf_roc_auc = auc(rf_fpr, rf_tpr)
````
